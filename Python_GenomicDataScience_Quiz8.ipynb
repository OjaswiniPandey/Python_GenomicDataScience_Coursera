```python
from Bio.Blast import NCBIWWW
```

```python
fasta_string = open ("C:\\Users\\ojasw\\Desktop\\Genomic_data_Science\\Python_For_Genomic_data_science\\mySeq.fa.txt").read()
```

```python
print (fasta_string)
```

```python
result_handle = NCBIWWW.qblast("blastn", "nt", fasta_string)  
```

```python
from Bio.Blast import NCBIXML
```

```python
blast_record = NCBIXML.read(result_handle)
```

```python
len(blast_record.alignments)
```

```python
E_value_threshold = 0.01
for alignments in blast_record.alignments:
    for hsp in alignments.hsps:
        if hsp.expect < E_value_threshold:
            print ('*****Alignment*****')
            print ('sequence :',alignments.title)
            print ('length:',alignments.length)
            print ('e value:', hsp.expect)
            print (hsp.query)
            print (hsp.match)
            print (hsp.sbjct)            
```
